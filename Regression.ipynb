```python
import pandas as pd
import statsmodels.formula.api as sm
import seaborn as sns
df = pd.read_pickle('./full_data.pkl')
df = df.set_index('dates')

df['populated'] = pd.notna(df.president).astype(int)

grouped = df.groupby('president')

grouped.get_group('Barack Obama')
```

Check to see if whether there is a speech of not affects the stock market

```python
model = sm.ols(formula = 'Dow_Change ~ populated', data = df).fit()

model.summary()
```

We can see that there is no affect so we can ignore that from now on

```python
model = sm.ols(formula = 'Dow_Change ~ compound', data = df).fit()

model.summary()
```

```python
sns.regplot(x = "compound", y = 'Dow_Change', data = df)
```

```python
model = sm.ols(formula = 'Dow_Change ~ positive', data = df).fit()

model.summary()
```

```python
sns.regplot(x = "positive", y = 'Dow_Change', data = df)
```

```python
model = sm.ols(formula = 'Dow_Change ~ negative', data = df).fit()

model.summary()
```

```python
sns.regplot(x = "negative", y = 'Dow_Change', data = df)
```
