```python
import time
import re
from bs4 import BeautifulSoup #use beautifulsoup to scrape data
from selenium import webdriver
import pandas as pd



url = "https://millercenter.org/the-presidency/presidential-speeches"


browser = webdriver.Chrome() #change this line depending on your system
browser.get(url) #go to the website 

time.sleep(2)


date_select = browser.find_element_by_xpath('/html/body/div[2]/div/main/div[2]/div/div[2]/article/div/\
div[2]/div/div/div/form/div[3]/ul/li[2]/a')

date_select.click()


date_pick_from = browser.find_element_by_xpath('//*[@id="edit-field-speech-date-value-min"]')
#ask for input for date (hardcoded for now)

date_pick_from.click()
date_pick_from.send_keys('01/01/2018')

time.sleep(1)


#date to is necessary because the website sucks. It won't disable scrolling to the bottom feature
#if it doesn't have both dates 
date_pick_to = browser.find_element_by_xpath('//*[@id="edit-field-speech-date-value-max"]')

date_pick_to.click()
date_pick_to.send_keys('03/19/2018')





time.sleep(4)

SCROLL_PAUSE_TIME = 0.8


#slight changes to be made!!

# Get scroll height
last_height = browser.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = browser.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height


time.sleep(2)
soup = BeautifulSoup(browser.page_source, 'lxml')    
```
